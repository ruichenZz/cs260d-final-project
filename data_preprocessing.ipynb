```python
import os
import re
import numpy as np

# Specify the directory containing the .npz files
subset_dir = "./subset"  # Update this to your actual directory path

# Initialize a set to store the union of all arrays
union_set = set()

# Function to check if the file name contains a number greater than 100
def contains_large_number(file_name):
    numbers = re.findall(r'\d+', file_name)  # Extract all numbers from the file name
    return any(int(num) > 100 for num in numbers)  # Check if any number is greater than 100

# Loop through all .npz files in the directory
for file_name in os.listdir(subset_dir):
    if file_name.endswith(".npz"):
        file_path = os.path.join(subset_dir, file_name)
        data = np.load(file_path)
        print(data['subset'].shape)
        # Extract and add the arrays to the union set
        union_set.update(data['subset'])

# Convert the union set to a sorted array (optional)
union_array = np.array(sorted(union_set))
```

```python
union_array.shape
```

```python
import pandas as pd

# Convert the NumPy array to a DataFrame
union_df = pd.DataFrame(union_array, columns=["id"])
output_csv_path = "./subset/union_of_ids_all.csv"  # Update the desired output path
union_df.to_csv(output_csv_path, index=False)
```

```python
# Load the original dataset
original_dataset_path = "./jigsaw-dataset/train.csv"  # Update with the actual path
original_dataset = pd.read_csv(original_dataset_path)

# Load the subset IDs (as row indices)
subset_ids_path = "./subset/union_of_ids.csv"  # Update with the actual path
subset_ids = pd.read_csv(subset_ids_path)  # Assuming no header in the subset file

# Ensure subset IDs are integers (row indices)
subset_ids["id"] = subset_ids["id"].astype(int)

# Slice the original dataset using subset indices
filtered_dataset = original_dataset.iloc[subset_ids["id"].values]

# Save the sliced dataset to a new CSV
output_sliced_path = "./sliced_dataset.csv"
filtered_dataset.to_csv(output_sliced_path, index=False)
```
